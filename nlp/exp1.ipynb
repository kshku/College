```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
def preprocess(text):
    tokens = nltk.word_tokenize(text)
    filtered = [word for word in tokens if word.isalnum()]
    validate = [word for word in filtered if re.match('[A-Za-z]+$', word)]
    words = [word for word in validate if word not in stopwords.words()]
    stemmer = PorterStemmer()
    words = [stemmer.stem(word) for word in words]
    return words
```

```python
text = "Natural Language Processing (NLP) is a fascinating field of Artificial Intelligence"
```

```python
preprocess(text)
```
