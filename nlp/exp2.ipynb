```python
import nltk
from nltk import word_tokenize, ngrams
from collections import Counter
```

```python
text = "I love programming. Programming is ful. I love coding and programming."
```

```python
words = word_tokenize(text.lower())
```

```python
def get_prob(n):
    ng = list(ngrams(words, n))
    c = Counter(ng)
    total = len(ng)
    return {k: c[k] / total for k in c}
```

```python
uni = get_prob(1)
bi = get_prob(2)
tri = get_prob(3)
```
